#Gold Layer

Objective: Develop a Gold Layer for Data Analytics

```python
df_silver=spark.table("workspace.silver.airbnb_listings_amsterdam")
```

Neighbourhoods KPI

```python
from pyspark.sql.functions import avg, count, sum,round

df_gold_neighbourhood = (
    df_silver.groupBy("neighbourhood")
    .agg(
        count("*").alias("total_listings"),
        round(avg("price"), 2).alias("avg_price"),
        round(avg("availability_365"), 2).alias("avg_availability"),
        round(sum("number_of_reviews"), 2).alias("total_reviews")
    )
)

display(df_gold_neighbourhood)
```

```python
df_gold_neighbourhood.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.airbnb_kpi_neighbourhood")
```

Room Type KPI

```python
from pyspark.sql.functions import avg, count, round

df_gold_room_type = (
    df_silver.groupBy("room_type")
    .agg(
        count("*").alias("total_listings"),
        round(avg("price"), 2).alias("avg_price"),
        round(avg("minimum_nights"), 2).alias("avg_minimum_nights"),
        round(avg("availability_365"), 2).alias("avg_availability")
    )
)

display(df_gold_room_type)
```

```python
df_gold_room_type.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.airbnb_kpi_room_type")
```

Hosts KPI

```python
from pyspark.sql.functions import count, avg,sum, round,col

df_gold_hosts = (
    df_silver.groupBy("host_id", "host_name")
    .agg(
        count("*").alias("total_listings"),
        round(avg("price"), 2).alias("avg_price"),
        round(sum("number_of_reviews"), 2).alias("total_reviews")
    )
    .orderBy(col("total_listings").desc())
)

display(df_gold_hosts)
```

```python
df_gold_hosts.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.airbnb_top_hosts")
```
